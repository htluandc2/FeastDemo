```python
import pandas as pd
import feast
```

```python
load_df = pd.read_parquet('credit_scoring/feature_repo/raw_data/loan_table.parquet')
load_df
```

```python
fs = feast.FeatureStore(repo_path='credit_scoring/feature_repo/')
feature_service_v1 = fs.get_feature_service("features_group_v1")
feature_service_v2 = fs.get_feature_service("features_group_v2")
```

```python
training_df = fs.get_historical_features(
    entity_df=load_df,
    features=feature_service_v1
).to_df()
training_df2= fs.get_historical_features(
    entity_df=load_df,
    features=feature_service_v2
).to_df()
```
